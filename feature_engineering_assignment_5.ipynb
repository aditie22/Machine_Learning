Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

xi={-4,-3,-2,-1,0,1,2,3,4}
yi={16,9,4,1,0,1,4,9,16}

To calculate the Pearson correlation coefficient for the given dataset, you can follow these steps:

Calculate the mean of both datasets.
Calculate the difference between each data point and its corresponding mean for both datasets.
Calculate the product of the differences for each pair of data points.
Sum up the products obtained in step 3.
Calculate the square root of the sum of the squared differences between each data point and its corresponding mean for both datasets.
Divide the sum obtained in step 4 by the value obtained in step 5 to get the Pearson correlation coefficient.
Let's calculate it step by step:

xi: {-4, -3, -2, -1, 0, 1, 2, 3, 4}
yi: {16, 9, 4, 1, 0, 1, 4, 9, 16}

Calculate the mean of both datasets:
Mean(xi) = (-4 - 3 - 2 - 1 + 0 + 1 + 2 + 3 + 4) / 9 = 0
Mean(yi) = (16 + 9 + 4 + 1 + 0 + 1 + 4 + 9 + 16) / 9 = 8

Calculate the differences from the mean:
for xi: {-4, -3, -2, -1, 0, 1, 2, 3, 4}
differences: {-4, -3, -2, -1, 0, 1, 2, 3, 4}

for yi: {16, 9, 4, 1, 0, 1, 4, 9, 16}
differences: {8, 1, -4, -7, -8, -7, -4, 1, 8}

Calculate the product of differences for each pair of data points and sum them up:
sum(xi * yi) = (-4 * 8) + (-3 * 1) + (-2 * -4) + (-1 * -7) + (0 * -8) + (1 * -7) + (2 * -4) + (3 * 1) + (4 * 8) = 0

Calculate the squared differences and sum them up for both datasets:
sum(xi^2) = (-4)^2 + (-3)^2 + (-2)^2 + (-1)^2 + 0^2 + 1^2 + 2^2 + 3^2 + 4^2 = 30
sum(yi^2) = 8^2 + 1^2 + (-4)^2 + (-7)^2 + (-8)^2 + (-7)^2 + (-4)^2 + 1^2 + 8^2 = 322

Calculate the square root of the sum of squared differences for both datasets:
sqrt(sum(xi^2)) = sqrt(30) ≈ 5.48
sqrt(sum(yi^2)) = sqrt(322) ≈ 17.94

Calculate the Pearson correlation coefficient:
Pearson correlation = sum(xi * yi) / (sqrt(sum(xi^2)) * sqrt(sum(yi^2)))
Pearson correlation = 0 / (5.48 * 17.94) ≈ 0

The Pearson correlation coefficient for the given dataset is approximately 0, indicating a very weak linear relationship between the two variables.
